```python
# Which team have win most matches 
w = pd.DataFrame({"Winner":df['winner']})
count_wins = w.value_counts()
print(count_wins)
```

```python
# Which team have win most matches 
sns.barplot(x=df['winner'].value_counts().head(10).index,
           y=df['winner'].value_counts().head(10).values,
           data = df)
plt.xticks(rotation= 70)
plt.title('Winners')
plt.xlabel('winner')
plt.ylabel('Teams')
plt.show()
```

```python
label = [X[0] for X in count_wins.keys()]
label
```

```python
# Most of wins
bar , ax = plt.subplots(figsize=(25,15))
ax = plt.pie(x = count_wins, autopct = "%.1f%%",labels = label)
plt.title("Most wins in IPL",fontsize=20)
plt.show()
```

```python
# Most win in  eliminator
sns.countplot(df['winner'][df['eliminator'] == 'Y'],data = df)
plt.title("Most win in eliminator",fontsize = 20)
plt.xticks(rotation = 90)
plt.show()
```

```python
# Which team choose bat field count
teams = df["toss_winner"].unique()
decision_making = pd.DataFrame([],columns= ['Toss Winner','Decision','Times'])
for id, element in enumerate(teams):
    team_bat = df[(df['toss_winner']== element) & (df['toss_decision']== 'bat')]
    team_field = df[(df['toss_winner']== element) & (df['toss_decision']== 'field')]
    
    decision_making = decision_making.append({'Toss Winner': element,'Decision': 'bat','Times' : team_bat['toss_winner'].count()}, ignore_index= True)
    decision_making = decision_making.append({'Toss Winner': element,'Decision': 'field','Times' : team_field['toss_winner'].count()}, ignore_index= True)
```
